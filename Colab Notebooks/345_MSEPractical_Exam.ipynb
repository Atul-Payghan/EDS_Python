```python
#1]Display the Highest price of car
import csv

def get_highest_price(content_toyota1.csv):
    highest_price = 0

    with open("/content/toyota1.csv", 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            price = float(row['Price'])
            if price > highest_price:
                highest_price = price

    return highest_price


csv_file = "/content/toyota1.csv"
highest_price = get_highest_price("/content/toyota1.csv")
print(f"The highest price of a car is: {highest_price}")
```
