```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [10,15,7,18,65,32,65]

plt.plot(x,y,marker = 'o')
plt.title("line Chart Example")
plt.xlabel("Time")
plt.ylabel("Dependent Variable")

plt.show()
```
